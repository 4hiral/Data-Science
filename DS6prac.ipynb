```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
sns.set(style="whitegrid")
```

```python
#Histogram
plt.figure(figsize=(6,4))
sns.histplot(df['total_bill'], kde=True, color="skyblue")
plt.title("Distribution of Total Bill")
plt.show()
```

```python
#Countplot
plt.figure(figsize=(6,4))
sns.countplot(x="day", data=df, palette="pastel")
plt.title("Count of Records by Day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex", palette="Set2")
plt.title("Total Bill vs Tip by Gender")
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x="day", y="total_bill", data=df, palette="muted")
plt.title("Boxplot of Total Bill by Day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
